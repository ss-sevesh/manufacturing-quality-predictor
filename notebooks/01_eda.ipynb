# Exploratory Data Analysis
## Manufacturing Quality Score Prediction

This notebook explores the synthetic manufacturing dataset to understand:
- Feature distributions and outliers
- Correlations between features and quality score
- Data quality checks

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from src.data.generate_data import generate_manufacturing_data
from src.data.validate import validate_data
from src.visualization.plots import plot_feature_distributions, plot_correlation_matrix

%matplotlib inline
plt.style.use('seaborn-v0_8-whitegrid')
pd.set_option('display.max_columns', None)
```

```python
# Generate or load data
df = generate_manufacturing_data(num_samples=10000, seed=42)
print(f'Shape: {df.shape}')
df.head()
```

```python
# Summary statistics
df.describe().round(3)
```

```python
# Data validation
validate_data(df)
print('All validations passed')
```

```python
# Feature distributions
fig = plot_feature_distributions(df)
plt.show()
```

```python
# Correlation matrix
fig = plot_correlation_matrix(df)
plt.show()
```

```python
# Quality score distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].hist(df['quality_score'], bins=50, edgecolor='black', alpha=0.7)
axes[0].set_title('Quality Score Distribution')
axes[0].set_xlabel('Quality Score')

# Top correlations with quality score
corr_with_target = df.corr()['quality_score'].drop('quality_score').sort_values()
corr_with_target.plot.barh(ax=axes[1])
axes[1].set_title('Feature Correlation with Quality Score')
plt.tight_layout()
plt.show()
```

```python
# Interactive scatter: top feature vs quality
top_feature = corr_with_target.abs().idxmax()
fig = px.scatter(df, x=top_feature, y='quality_score', opacity=0.3,
                 title=f'{top_feature} vs Quality Score')
fig.show()
```
